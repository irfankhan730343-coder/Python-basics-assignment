Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
a=1
while a<=20:
    print(a)
    a=a+1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
#write answer here

n=int(input("Enter a number  :"))
total =0
i=1
while i<=n:
  total=total+i
  i=i+1
print("sum from 1 to",n,"is:",total)

```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
num=int(input("Enter a number"))
count=0

if num==0:
  count=1
else:
  while num>0:
    num = num//10
    count=count+1
    print("Number of digit :",count)
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
#write answer here
secret_number = 7
guess = None

while guess != secret_number:
    guess = int(input("Guess the number: "))

    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        print("Correct! You guessed the number.")
```
