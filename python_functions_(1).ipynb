## Question 1: Square of a Number

**Problem:**
Create a function `square_number` that takes one number and returns its square.

**Approach:**
- Define a function with one parameter
- Use `return` to send back the squared value

```python
#write your answer here

def square_number(n):
  return n**2

square_number(5)
```

## Question 2: Check Even or Odd

**Problem:**
Create a function `check_even_odd` that returns `Even` or `Odd`.

**Approach:**
- Use modulus operator `%`
- Apply conditional logic inside function

```python
#write your answer here
def check_even_odd(n):
  for i in n:
    if i %2==0:
      print("even")

    else:
      print("odd")


check_even_odd([1,2,3,4,5,6,7,8,9])
```

## Question 3: Count Vowels in a String

**Problem:**
Create a function `count_vowels` that counts vowels in a string.

**Approach:**
- Loop through each character
- Check if character is a vowel

```python
#write your answer here
def count_vowels(n):
  vowels= "aeiou"
  for i in vowels:
    if i in n:
      print(i)


count_vowels("irfan")
```

## Question 4: Find Maximum of Three Numbers

**Problem:**
Create a function `find_max` that returns the largest of three numbers without using `max()`.

**Approach:**
- Compare numbers using conditional statements

```python
def find_max(a, b, c):
  if a >= b and a >= c:
    return a
  elif b >= a and b >= c:
    return b
  else:
    return c

# Test cases
print(find_max(10, 5, 8))
print(find_max(3, 12, 7))
print(find_max(1, 4, 9))
```
